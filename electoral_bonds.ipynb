```python
import tabula
```

```python
filename = 'encashment_details.pdf'
```

```python
tables = tabula.read_pdf(filename, pages='all', pandas_options={'header': None})
```

```python
cols = tables[0].values.tolist()[0]
```

```python
tables[0] = tables[0].iloc[1:]
```

```python
import pandas as pd
df = pd.concat(tables)
```

```python
df.columns = cols
```

```python
df.head()
```

```python
df = df.rename(columns={'Date of\rEncashment': 'date', 'Name of the Political Party': 'name', 'Denomination': 'amount'})
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df['name'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.hist(df['name'])
plt.xticks(rotation=90)
```

```python
name_data = df['name'].value_counts().sort_values(ascending=False)
sns.barplot(name_data)
plt.xticks(fontsize = 8, rotation=90)
plt.xlabel('Name of the Party')
plt.ylabel('Number of electoral bonds encashed')
```

```python
df.head()
```

```python
df['amount'] = df['amount'].map(lambda x: int(x.replace(',', '')))
```

```python
df.head()
```

```python
amount_data = df.groupby('name')['amount'].sum()
```

```python
amount_data
```

```python
amount_data.apply(lambda x: x / 1e7)
```

```python
sns.barplot(amount_data.sort_values(ascending=False).apply(lambda x: x / 1e7))
plt.xticks(fontsize = 8, rotation=90)
plt.xlabel('Name of the Party')
plt.ylabel('Amount encashed with electoral bonds (in Cr)')
```

```python
df['name'].value_counts()
```

```python
orgs_of_interest = ['BHARTIYA JANTA PARTY', 'ALL INDIA TRINAMOOL CONGRESS', 'PRESIDENT, ALL INDIA CONGRESS COMMITTEE', ]
filtered_df = df[df['name'].isin(orgs_of_interest)]
grouped_df = filtered_df.groupby(by = ['date', 'name'])['amount'].sum().reset_index()
grouped_df
```

```python
grouped_df['amount'] = grouped_df['amount'].apply(lambda x: x / 1e7)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))

# Plot the data
sns.lineplot(x='date', y='amount', data=grouped_df, hue='name', palette='husl')

# Set labels and title
plt.title('Time Trends in Donations')
plt.xlabel('Year')
plt.ylabel('Total Amount Donated')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import tabula
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

filename = 'purchaser_details.pdf'
purchaser_tables = tabula.read_pdf(filename, pages='all', pandas_options={'header': None})
purchaser_cols = purchaser_tables[0].values.tolist()[0]
purchaser_tables[0] = purchaser_tables[0].iloc[1:]
purchaser_df = pd.concat(purchaser_tables)
purchaser_df.columns = purchaser_cols
```

```python
purchaser_df.head()
```

```python
purchaser_df = purchaser_df.rename(columns = {'Date of Purchase': 'date', 'Purchaser Name': 'name', 'Denomination': 'amount'})
```

```python
purchaser_df.head()
```

```python
purchaser_df['date'] = pd.to_datetime(purchaser_df['date'])
```

```python
purchaser_df['amount'] = purchaser_df['amount'].map(lambda x: int(x.replace(',', '')))
```

```python
purchaser_df.head()
```

```python
name_data = purchaser_df['name'].value_counts().sort_values(ascending=False).head(15)
sns.barplot(name_data)
plt.xticks(fontsize = 8, rotation=90)
plt.xlabel('Name of the Donator')
plt.ylabel('Number of electoral bonds donated')
```

```python
amount_data = purchaser_df.groupby('name')['amount'].sum()
```

```python
amount_data.head()
```

```python
sns.barplot(amount_data.sort_values(ascending=False).apply(lambda x: x / 1e7).head(20))
plt.xticks(fontsize = 8, rotation=90)
plt.xlabel('Name of the Donator')
plt.ylabel('Amount donated with electoral bonds (in Cr)')
```
